```python
# Dados para treinamento

import sqlite3
import pandas as pd

conn = sqlite3.connect("database.db")

consulta_atividade = """
  SELECT *
  FROM flight_activity fa LEFT JOIN flight_loyalty_history flh
          ON (fa.loyalty_number = flh.loyalty_number)
"""

df_atividade = pd.read_sql_query(consulta_atividade, conn)

# Selecionar as linhas que contém apenas dados numéricos
colunas  = ["year", "month", "flights_booked", "flights_with_companions",
            "total_flights", "distance", "points_accumulated", "salary", "clv"]

df_colunas_selecionadas = df_atividade.loc[: , colunas]

# Remover todas as linhas que contém alguma coluna com dados faltantes
df_treinamento = df_colunas_selecionadas.dropna()

# Verificar se os dados não tem dados faltantes
df_treinamento.isna().sum()
```

# Exercícios

```python
# 1. Qual o valor da soma total da distância percorrida pelos voos registrados
# na planilha de dados?

df_atividade.loc[:, 'distance'].sum()
```

```python
# 2. Qual o salário médio dos passageiros?

df_atividade.loc[:, 'salary'].mean()
```

```python
# 3. Qual o valor total de pontos acumulados?

df_atividade.loc[:, 'points_accumulated'].sum()
```

```python
# 4. Selecione linhas com valores específicos em uma coluna: Selecione apenas as
# linhas em que a coluna 'year' seja igual a 2023 e mostre as colunas 'month',
# 'total_flights' e 'clv'.

linhas = df_atividade['year'] == 2023
colunas = ['month', 'total_flights', 'clv']

df_selecionado = df_atividade.loc[linhas, colunas]
print(df_selecionado)
```

```python
# 5. Filtrar linhas com base em uma condição múltipla:Selecione apenas as c
# olunas 'year', 'salary' e 'points_accumulated' do DataFrame df_atividade.

colunas = ['year', 'salary', 'points_accumulated']

df_selecionado = df_atividade.loc[:, colunas]
print(df_selecionado)
display(df_selecionado)
```

```python
# 6. Selecionar linhas com base em um intervalo de valores: Mostre todas as
# colunas do DataFrame df_atividade para as primeiras 10 linhas.

df_selecionado = df_atividade.head(10)
print(df_selecionado)
display(df_selecionado)
```
